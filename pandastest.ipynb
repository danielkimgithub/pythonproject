```python
usa = oo[oo.NOC == 'USA']
usa = usa.groupby(['Medal', 'Athlete', 'Edition']).size().unstack('Medal', fill_value = 0)
usa['Total'] = usa['Gold'] + usa['Silver'] + usa['Bronze']
usa.reset_index(inplace=True)
```

```python
for year, group in usa.groupby('Edition'):
    print(group.sort_values('Total', ascending = False)[:1])
```

```python
tu = [group.sort_values('Total', ascending = False)[:1] for year, group in usa.groupby('Edition')]
tu
top = pd.DataFrame()
for i in tu:
    top = top.append(i)
top
```
